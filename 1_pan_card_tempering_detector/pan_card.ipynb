# PAN Card Tampering Detection

##### Tujuan proyek ini adalah untuk mendeteksi pemalsuan kartu PAN menggunakan visi komputer. Proyek ini akan membantu berbagai organisasi dalam mendeteksi apakah identitas, yaitu kartu PAN yang diberikan kepada mereka oleh karyawan, pelanggan, atau siapa pun, adalah asli atau tidak.

##### Untuk proyek ini, kami akan menghitung kemiripan struktural antara kartu PAN asli dan kartu PAN yang diunggah oleh pengguna.

```python
# import the necessary packages
from skimage.metrics import structural_similarity
import imutils
import cv2
from PIL import Image
import requests
```

```python
# Open image and display
original = Image.open(requests.get('https://www.thestatesman.com/wp-content/uploads/2019/07/pan-card.jpg', stream=True).raw)
tampered = Image.open(requests.get('https://assets1.cleartax-cdn.com/s/img/20170526124335/Pan4.png', stream=True).raw) 
```

##### Memuat gambar asli dan gambar yang disediakan pengguna.

```python
# The file format of the source file.
print("Original image format : ",original.format) 
print("Tampered image format : ",tampered.format)

# Image size, in pixels. The size is given as a 2-tuple (width, height).
print("Original image size : ",original.size) 
print("Tampered image size : ",tampered.size) 
```

##### Mengubah format gambar yang telah dimodifikasi agar mirip dengan format gambar asli.

```python
# Resize Image
original = original.resize((250, 160))
print(original.size)
original.save('pan_card_tampering/image/original.png')#Save image
tampered = tampered.resize((250,160))
print(tampered.size)
tampered.save('pan_card_tampering/image/tampered.png')#Saves image
```
